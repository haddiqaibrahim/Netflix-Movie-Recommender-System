#Model Building

```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Model Building - MovieRecommender Class
class MovieRecommender:
    def __init__(self, dataset_path):
        self.dataset = pd.read_csv(dataset_path)
        self.tfidf_matrix = None
        self.similarity_matrix = None
        self.vectorizer = TfidfVectorizer(stop_words='english')
        self.preprocess_data()
        self.build_model()

    def preprocess_data(self):
        for col in ['director', 'cast', 'country', 'date_added', 'rating', 'description', 'listed_in']:
            self.dataset[col] = self.dataset[col].fillna('Unknown')
        self.dataset['combined_features'] = self.dataset['listed_in'] + ' ' + self.dataset['description']

    def build_model(self):
        self.tfidf_matrix = self.vectorizer.fit_transform(self.dataset['combined_features'])
        self.similarity_matrix = cosine_similarity(self.tfidf_matrix)

    def recommend(self, title, num_recommendations=5):
        title = title.lower()
        titles = self.dataset['title'].str.lower()
        if title not in titles.values:
            return f"No movie found with title: '{title}'"
        idx = titles[titles == title].index[0]
        similarity_scores = list(enumerate(self.similarity_matrix[idx]))
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
        recommended_indices = [i[0] for i in similarity_scores[1:num_recommendations+1]]
        return self.dataset.iloc[recommended_indices][['title', 'listed_in', 'description']]
```

#Model Testing + Evaluation

```python
# Step 3: Model Testing - Recommend Movies
recommender = MovieRecommender("netflix_titles.csv")
test_title = "Sherlock"
recommended = recommender.recommend(test_title, num_recommendations=5)

print(f"Top 5 recommendations for '{test_title}':\n")
print(recommended)
```

```python
# Step 4: Define Evaluation Metrics
ground_truth_similar_movies = {
    'sherlock': ['Top Boy', 'Case Closed', 'Criminal: UK', 'Paranoid', 'Luther']
}

def precision_at_k(recommended_df, relevant_titles, k=5):
    recommended_titles = recommended_df['title'].str.lower().tolist()
    relevant_titles = [title.lower() for title in relevant_titles]
    hits = sum([1 for title in recommended_titles if title in relevant_titles])
    return hits / k

def recall_at_k(recommended_df, relevant_titles, k=5):
    recommended_titles = recommended_df['title'].str.lower().tolist()
    relevant_titles = [title.lower() for title in relevant_titles]
    hits = sum([1 for title in recommended_titles if title in relevant_titles])
    return hits / len(relevant_titles)

def f1_score_at_k(precision, recall):
    if precision + recall == 0:
        return 0
    return 2 * (precision * recall) / (precision + recall)

def diversity_at_k(recommended_df, k=5):
    recommended_titles = recommended_df['title'].tolist()
    unique_titles = len(set(recommended_titles))
    return unique_titles / k
```

```python
# Step 5: Calculate Evaluation Metrics
if test_title.lower() in ground_truth_similar_movies:
    precision = precision_at_k(recommended, ground_truth_similar_movies[test_title.lower()], k=5)
    recall = recall_at_k(recommended, ground_truth_similar_movies[test_title.lower()], k=5)
    f1_score = f1_score_at_k(precision, recall)
    diversity = diversity_at_k(recommended, k=5)

    print(f"\nPrecision@5 for '{test_title}': {precision:.2f}")
    print(f"Recall@5 for '{test_title}': {recall:.2f}")
    print(f"F1 Score@5 for '{test_title}': {f1_score:.2f}")
    print(f"Diversity@5 for '{test_title}': {diversity:.2f}")
else:
    print("\nNo ground truth available for evaluation.")
```

```python
# Step 6: Visualize Evaluation Metrics
metrics = ['Precision@5', 'Recall@5', 'F1 Score@5', 'Diversity@5']
scores = [precision, recall, f1_score, diversity]

plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, scores, color=['#4caf50', '#2196f3', '#ff9800', '#9c27b0'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height:.2f}', ha='center', va='bottom')

plt.title('Evaluation Metrics for Movie Recommender System')
plt.ylim(0, 1.1)
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#  **Summary of Evaluation Metrics**

This bar chart illustrates the evaluation of the **Movie Recommender System** based on four key metrics for the movie **"Sherlock"**:

- **Precision@5 (0.80)**: Out of the 5 recommended movies, 80% were actually relevant. This means the model is good at suggesting accurate recommendations.
- **Recall@5 (0.80)**: 80% of all relevant movies (from ground truth) were successfully recommended. This shows the model is not missing many good options.
- **F1 Score@5 (0.80)**: A balance between precision and recall, indicating consistent performance across both accuracy and coverage.
- **Diversity@5 (1.00)**: All 5 recommendations were unique, showing that the model suggests a variety of options rather than similar or repetitive titles.

These metrics help evaluate the **effectiveness and quality** of the recommender system  balancing accuracy with variety to enhance user experience.
